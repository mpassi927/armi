```python
#Divisons notre base de donnÃ©e en deux (une partie contenant les objets et une autre les numÃ©riques)
cat_data=[]
num_data=[]
for i,j in enumerate(all_data.dtypes):
    if j==object:
        cat_data.append(all_data.iloc[:,i])
    else:
        num_data.append(all_data.iloc[:,i])
cat_data=pd.DataFrame(cat_data).transpose()
num_data=pd.DataFrame(num_data).transpose()
```

```python
#Pour les variables catÃ©gories(cat_data), nous allons remplacer les valeurs manquantes par les valeurs qui se repetent le plus
cat_data=cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum().any()
```

```python
#Pour les variables numÃ©riques(num_data), nous allons remplacer les valeurs manquantes par les valeurs qui sont proches de cette derniere
num_data.fillna(method="bfill", inplace=True)
num_data.isnull().sum().any()
```

```python
#Transformer la colonne target
target_values={"Y":1,"N":0}
target=cat_data["Loan_Status"]
cat_data.drop("Loan_Status",axis=1, inplace=True)
target=target.map(target_values)
target
```

```python
#Remplacer les catÃ©gories par les valueurs numÃ©riques 1,2,3,...
le=LabelEncoder()
for c in cat_data:
    cat_data[c]=le.fit_transform(cat_data[c])
cat_data
```

```python
#Supprimer la colonne Loan_ID 
cat_data.drop("Loan_ID",axis=1, inplace=True)
```

```python
#Concatener les deux datas cat_data et num_data et spÃ©cifier la colonne target
X=pd.concat([cat_data,num_data], axis=1)
Y=target
```

```python
#diviser la base de donnÃ©e en une base de donnÃ©e test et d'entrainement
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X,Y):
    X_train,X_test=X.iloc[train],X.iloc[test]
    Y_train,Y_test=Y.iloc[train],Y.iloc[test]
    
print("X_train est ", X_train.shape)
print("X_test est ", X_test.shape)
print("Y_train est ", Y_train.shape)
print("Y_test est ", Y_test.shape)
```

```python
#On va utiliser trois algorithmes du machine learning: Logistic Regression, KNN, Decisiontree
models={
    "LogisticRegression":LogisticRegression(random_state=42),
    "KNeighborsClassifier":KNeighborsClassifier(),
    "DecisionTreeClassifier":DecisionTreeClassifier(max_depth=1,random_state=42)
}

#la fonction de prÃ©cision
def accu(y_true,y_pred,retu=False):
    acc=accuracy_score(y_true,y_pred)
    if retu:
        return acc
    else:
        print(f"la prÃ©cision du model est :{acc}")
        
#La fonction d'application des models
def train_test_eval(models,X_train,Y_train,X_test,Y_test):
    for name,model in models.items():
        print(name,":")
        model.fit(X_train,Y_train)
        accu(Y_test,model.predict(X_test))
        print("-"*30)
        
train_test_eval(models,X_train,Y_train,X_test,Y_test)
```
